```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

```python
Total_Player_Count = purchase_data["SN"].drop_duplicates().count()
df_player = {'Total Player Count':[Total_Player_Count]}
pd.DataFrame(df_player)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Price_Average = purchase_data["Price"].mean()
Price_Average_Round = round(Price_Average,2)

Unique_Items = purchase_data['Item ID'].drop_duplicates().value_counts().sum()

Purchase_Number = purchase_data['Purchase ID'].count()

Total_Rev = purchase_data['Price'].sum()


Sum_df = {'Number of Unique Items':[Unique_Items],'Average Price':["$" + str(Price_Average_Round)],'Number of Purchase':[Purchase_Number],'Total Revenue':['$' + str(Total_Rev)]}
pd.DataFrame(Sum_df)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]

group_names = ["0-9","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

demo_total_2 = purchase_data.drop_duplicates(['SN'], keep='last')
demo_total_3 = demo_total_2.groupby("Age").count()
d_t = demo_total_3["Purchase ID"]
#d_t

pct_pl = round(((d_t/Total_Player_Count)*100),2)
#pct_pl

demo_df = pd.DataFrame({"Total Count":d_t,"Percentage of Players":pct_pl})
demo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pur_count_spend = purchase_data.groupby("SN").count()
pur_count_top = pur_count_spend["Purchase ID"]
#pur_count_top

pur_count_price = purchase_data.groupby("SN").mean()
pur_count_avg = round(pur_count_price["Price"],2)
#pur_count_avg

pur_count_sum = purchase_data.groupby("SN").sum()
pur_count_total = round(pur_count_sum["Price"],2)
#pur_count_total

top_spenders = pd.DataFrame({"Purchase Count":pur_count_top,"Average Purchase Price":pur_count_avg,"Total Purchase Value":pur_count_total})

top_spenders = top_spenders.sort_values(by=['Total Purchase Value'], ascending=False)

top_spenders.head()
                              
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pur_count_pop = purchase_data.groupby(["Item ID","Item Name"]).count()
pur_count_pop_total = pur_count_pop["Purchase ID"]
#pur_count_pop_total

item_price_pop = purchase_data.groupby(["Item ID","Item Name"]).mean()
item_price = item_price_pop['Price']


total_pur_value = purchase_data.groupby(["Item ID","Item Name"]).sum()
total_pur_value_pop = total_pur_value["Price"]
#total_pur_value_pop

most_popular = pd.DataFrame({"Purchase Count":pur_count_pop_total,"Item Price":item_price,"Total Purchase Value":total_pur_value_pop})

most_popular = most_popular.sort_values(by=['Purchase Count'], ascending=False)

most_popular.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = most_popular.sort_values(by=['Total Purchase Value'], ascending=False)

most_profitable.head()
```
